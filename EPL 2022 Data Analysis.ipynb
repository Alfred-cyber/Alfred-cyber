```python
epl_df['Average mins Per Match'] = (epl_df['Min']/epl_df['Apps']).astype(int)
epl_df['Goals Per Match'] = (epl_df['G']/epl_df['Apps']).astype(float)
epl_df.head()
```

```python
Total_goals_scored = epl_df['G'].sum()
print(Total_goals_scored)
```

```python
epl_df['Team'].value_counts().nlargest(5).plot(kind = 'bar', color = sns.color_palette("viridis"))
```

```python
epl_df['Team'].value_counts().nsmallest(5).plot(kind = 'bar', color = sns.color_palette("viridis"))
```

```python
#top 10 assists
top_10_assists = epl_df[['Player', 'Team', 'A', 'Apps']].nlargest(n = 10, columns = 'A')
top_10_assists
```

```python
#total goals by each club
Goals_by_clubs = pd.DataFrame(epl_df.groupby('Team', as_index=False)['G'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Team', y='G', data = Goals_by_clubs.sort_values(by="G")) #palette='rocket'
ax.set_xlabel("Club", fontsize = 30)
ax.set_xlabel("Goals", fontsize = 20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Goals', fontsize = 15)
```

```python
#most goals by players
top_10_goals = epl_df[['Player', 'Team', 'G', 'Apps']].nlargest(n = 10, columns = 'G')
top_10_goals
```

```python
#Goals per match
top_10_goals_permatch = epl_df[['Player', 'Goals Per Match', 'G', 'Apps']].nlargest(n = 10, columns = 'Goals Per Match')
top_10_goals_permatch
```

```python
#Pie chart of Goals with assist and without assist
plt.figure(figsize=(14,7))
assists = epl_df['A'].sum()
data = [(Total_goals_scored-assists), assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, autopct='%.0f%%')
plt.show
```
